```python
years = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]
data1 = [-21.9, -26, -27.1, -29.1, -30.1, -22.7, -18.4, -20.5, -15.8, -11.9]
data2 = [-10.2, 9.3, 3.9, -0.5, 9.2, 0, 3, 1.3, 11.5, -2.3]
data3 = [18.7, 11.8, 12.2, 17.2, 16.6, 11.1, 10.6, 7.6, 7.1, 15.5]

plt.plot(years, data1, label='account balance')
plt.plot(years, data2, label='overall balance')
plt.plot(years, data3, label='debt')

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Data')
plt.title('Account balance and debt service')
plt.legend()

# Displaying the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

From the given data it is aware that the balance of payment of a country is negative throughout indicating overdependency on import and an imbalance economy. import has decreased dractically during pandemic the consequence was burdansome to the nation but balance of payment shows robust, experiencing some growth but gradually its growth deminished after pandemic was lifted due to economic recovery efforts. The government needs to assist export-oriented businesses, cut spending, and diversify its revenue streams in order to achieve budgetary stability and economic resilience. Reducing dependency on imports and managing debt well are essential. Comprehensive policies have the power to attract foreign investment, improve trade performance, fortify the budgetary position, and ensure long-term economic growth and stability.
